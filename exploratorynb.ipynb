### Exploration of `movie_basics` and `movie_ratings

```python
cmovie_basics = movie_basics.copy() # copy movie_basics to new dataset 'cmovie_basics'
cmovie_basics
```

While exploring movie_basics we see that some of the films are categorized as multiple genres so we are going to need to separate the genres to get a more accurate account of genre ratings.

```python
cmovie_basics['genres'] = cmovie_basics['genres'].map(str) # cast genres column as string object
```

```python
cmovie_basics['genres'] = cmovie_basics['genres'].str.split(',') # split each element in genres column
```

```python
cmovie_basics['genres'].explode()
cmovie_basics = cmovie_basics.explode('genres').reset_index(drop=True)   # explode each genre
```

```python
cmovie_basics.head()
```

Now we have the movie separated into a different row for each genre that they are categorized in (e.g. Sunghursh was categorized as action, crime, and, drama).

```python
cmovie_basics.info()  # Looking for Non-Null Counts
```

```python
cmovie_ratings = movie_ratings.copy() # Copy movie_ratings to cmovie_ratings
```

```python
cmovie_ratings.info()
```

### Merge and plot `movie_ratings` and `movie_basics`

```python
ratings_basics_df = cmovie_ratings.merge(cmovie_basics, left_on='movie_id', right_on='movie_id') # merge movie_basics with
                                                                                                 # movie_ratings to get genres
                                                                                                 # averagerating
ratings_basics_df
```

We merge `movie_basics` and `movie_ratings` on `movie_id` into a single dataframe called `ratings_basics_df`.

```python
ratings_basics_df.info() # No non-null entries for 'averagerating' or 'genres' in our new dataframe
```

### Plot ratings and genres using merged database `ratings_basics_df`

Because genres with 'none' value makeup only .01% (804 of 129, 294 entries) of the database I will remove these columns.

```python
ratings_basics_df['start_year'].describe() # Checking which years this dataset contains to make sure the information is relevant
```

Now we can iterate over each unique genre of `ratings_basics_df` to find the average rating and put Genre and rating into their own dataframe to be plotted.

```python
genres_rating_df = ratings_basics_df[['genres', 'averagerating']].groupby(by='genres').mean('averagerating').sort_values('averagerating')
genres_rating_df.reset_index(inplace=True)
genres_rating_df
```

```python
ax = genres_rating_df.plot.barh(x = 'genres', y = 'averagerating', figsize = (18, 10))
ax.set_title('Average Rating of each Genre')
ax.set_xlabel('Average Rating (Out of 10)')
ax.set_ylabel('Genres');
```

This graph shows that 'Short' genre is rated the highest of all the data which raises a red flag because 'Short' is not a widely recognized category. We should look at the value count of the `ratings_basics` dataframe to see if this effects the mean of the categories.

```python
ratings_basics_df['genres'].value_counts()
```

```python
ratings_basics_df.loc[ratings_basics_df['genres'] == 'Short']
```

It does in fact look like 'Short' only has 1 count and the `averagerating` for that value is 8.8 which makes the graph look like Short is the most popular movie.

There are four categories (`Reality-TV`, `Adult`, `Game-Show`, and `Short`) whose value counts are low compared to the total dataset. In order to look at a more accurate depiction of the data, we should probably remove those lower value count ratings from the `genres_rating_df` using their respective indices.

```python
genres_rating_df = genres_rating_df.drop(genres_rating_df.index[[0, 16, 23, 25]])
genres_rating_df
```

```python
ax = genres_rating_df.plot.barh(x = 'genres', y = 'averagerating', figsize = (18, 10))
ax.set_title('Average Rating of each Genre')
ax.set_xlabel('Average Rating (Out of 10)')
ax.set_ylabel('Genres');
```

This horizontal bar chart looks much better with what looks like a more accurate depiction of the genres and their average ratings. The leading genre is 'Documentary' with an average rating of 7.33. We can actually see that 4 of the top 5 average rated categories are 'Documentary', 'News', 'Biography', and 'History'. These top categories could probably all be put under the umbrella category of history. I think that moving forward with a new production company, Microsoft should invest in making material that has some historical importance because it looks like the general public has a general interest in the category.

### Beginning to plot `budgets_genres`

This dataframe covered a large portion of time (1933-2018). So what I decided to make three separate plots. One plot with the full time period (1933-2018), the time preiod during the 20th century (1933-1999), and a final plot of this past 21st century so far (2000-2018). With this, I would be able to compare which genres were bringing production companies the most money.

```python
budgets_genres.info()  # Checking for non-null values in columns I need to graph (worldwide_gross, domestic_gross, ROI, genre)
```

```python
genres_ROI_1933_2018 = budgets_genres[['genre', 'ROI']].groupby(by='genre').mean('ROI').sort_values('ROI')
genres_ROI_1933_2018.reset_index(inplace=True)
genres_ROI_1933_2018
```

```python
bg_2000_2018 = budgets_genres.loc[budgets_genres['release_date'] >= '2000'] # Only to include genres after 2000
bg_2000_2018.head()
```

```python
genres_ROI_2000_2018 = bg_2000_2018[['genre', 'ROI']].groupby(by='genre').mean('ROI').sort_values('ROI')
genres_ROI_2000_2018.reset_index(inplace=True)
genres_ROI_2000_2018
```

```python
bg_1933_1999 = budgets_genres.loc[budgets_genres['release_date'] <= '1999'] # Only to include genres before 1999
bg_1933_1999.head()
```

```python
genres_ROI_1933_1999 = bg_1933_1999[['genre', 'ROI']].groupby(by='genre').mean('ROI').sort_values('ROI')
genres_ROI_1933_1999.reset_index(inplace=True)
genres_ROI_1933_1999
```

We can see that there are three missing rows for 'Sports and Fitness', 'Television', and 'Cult Movies'.

```python
fig, ax = plt.subplots(figsize=(18,10))
x = genres_list_4
y1 = genres_ROI_1933_2018['ROI']
y2 = genres_ROI_2000_2018['ROI']
y3 = genres_ROI_1933_1999['ROI']

ax.barh(x, y1, alpha = .8, color='darkblue', label = ('1933-2018'))
ax.barh(x, y2, alpha = .4, color= 'darkgreen', label=('2000-2018'))
#ax.barh(x, y3, alpha = .8, color = 'red', label =('1933-1999'))
ax.set_title('ROI by Genre')
ax.set_ylabel('Genres')
ax.set_xlabel('ROI (Return on Investment)')
ax.legend();
```

Looking at this graph, the most important color to note is the green because those are the most recent years ROI. Based on our other charts and graphs, 'Sci-Fi' and 'Fantasy' do not produce a lot of ROI probably because the production budget in those films are a lot higher due to special effects, possibly location, and the celebrity names that are drawn to those types of projects.

The 'Animation' genre does fall in the top 6 of ROI which makes sense because it is a cheaper genre to produce. It doesn't require flying any actors to particular destinations or paying particularly high salaries for voice actors and it is probably just a cheaper production option all around. In the previous charts the 'Animation' genre fell within the top 5 most popular or average rating of all the genres.
